```python
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
def mse(y_test,prediction):
    return np.mean((y_test-prediction)**2)
```

```python
X, Y = datasets.make_regression(n_samples=100,n_features=1,noise=30,random_state=4)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=123)

fig = plt.figure(figsize=(8,6))
plt.scatter(X,Y,color="b", marker='o',s=10)
plt.show()
```

```python
from LinearRegression import LinearRegression

l_reg = LinearRegression(epochs=1000,lr=0.05)
l_reg.fit(X_train,Y_train)
pred = l_reg.predict(X_test)
```

```python
lin_reg_mse = mse(Y_test,pred)
print('Mean Square Error',lin_reg_mse)

y_pred_line = l_reg.predict(X)

fig = plt.figure(figsize=(8,6))
plt.scatter(X_train,Y_train,color='b',s=10)
plt.scatter(X_test,Y_test,color='g',s=10)
plt.plot(X,y_pred_line,color='black',linewidth=2,label='Linear Regression Prediction')
```
